```python
x = open("accepted.csv", 'r')
c = csv.reader(x)
```

```python
for l in c:
    i, auth, title, abstract = l
    print('<p class="large text-muted">')
    print('<strong>' + title + '</strong>')
    print('<button class="btn btn-primary btn-xs" data-toggle="collapse" data-target="#pid' + i + '">Abstract</button>')
    print('<button class="btn btn-primary btn-xs" data-toggle="collapse" data-target="#bib' + i + '">BibTex</button>')
    print('<a href="papers/MLG2019_paper_' + i + '.pdf" target=_blank class="btn btn-primary btn-xs" role="button">PDF</a>')
    print('<button class="btn btn-primary btn-xs">S</button>')
    print('<br/>')
    print('<i>' + auth + '</i><br/>')
    print('<div id="pid' + i + '" class="collapse">')
    print('<strong>Abstract:</strong>' + abstract)
    print('<hr/>')
    print('</div>')
    print('<div id="bib' + i + '" class="collapse">')
    print('@inproceedings{mlg2019_' + i + ',<br/>')
    print('title={' + title + '},<br/>')
    print('author={' + auth + '},<br/>')
    print('booktitle={Proceedings of the 15th International Workshop on Mining and Learning with Graphs (MLG)},<br/>')
    print('year={2019}<br/>')
    print('}')
    print('<hr/>')
    print('</div>')
    print('</p>')
```
